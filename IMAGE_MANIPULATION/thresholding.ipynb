```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

path = "../dataset_images/20151101_145049.jpg"
image  = cv2.imread(path)
image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
ret , thresh1  =cv2.threshold(image , 120 , 255 , cv2.THRESH_BINARY)
plt.imshow(thresh1)
print(ret)
```

```python
ret , thresh2  =cv2.threshold(image , 120 , 255 , cv2.THRESH_BINARY_INV)
plt.imshow(thresh2)
```

```python
ret , thresh3  =cv2.threshold(image , 120 , 255 , cv2.THRESH_TRUNC)
plt.imshow(thresh3)
```

```python
ret , thresh4  =cv2.threshold(image , 120 , 255 , cv2.THRESH_TOZERO)
plt.imshow(thresh4)
```

```python
page = "../dataset_images/page.jpg"
page = cv2.imread(page)
plt.imshow(page)
```

```python
gray = cv2.cvtColor(page , cv2.COLOR_BGR2GRAY)
plt.imshow(gray);
```

# ADAPTIVE THRESHOLDING

```python
thresh_mean = cv2.adaptiveThreshold(gray , 255 , cv2.ADAPTIVE_THRESH_MEAN_C,
               cv2.THRESH_BINARY , 199 , 5)


plt.imshow(thresh_mean)
```

```python
thresh_guasssian  = cv2.adaptiveThreshold(gray , 255 , cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY , 199 , 5)
plt.imshow(thresh_guasssian)
```
