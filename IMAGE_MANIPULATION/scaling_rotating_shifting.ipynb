```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

path = "../dataset_images/20151101_145049.jpg"
image  = cv2.imread(path)
plt.imshow(image)
```

```python
image.shape
```

```python
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB) 
plt.imshow(image)
```

# ZOOMING IMAGE 

```python
height , width  = image.shape[:2]
```

```python
scale_factor = 3

zoomed_image = cv2.resize(src = image ,  
                          dsize= ( int (scale_factor*width ) , int(scale_factor*height)), 
                          interpolation= cv2.INTER_CUBIC)
plt.imshow(zoomed_image)
```

# ROTATING IMAGE

```python
plt.imshow(image)
```

```python
# image rotation 

center =(image.shape[1]  // 2 , image.shape[0] //2 )
angle = 30
scale = 1

rotated_matrix = cv2.getRotationMatrix2D(center , angle , scale)

rotated_image = cv2.warpAffine(image , rotated_matrix ,(image.shape[1] , image.shape[0]))

plt.imshow(rotated_image)
```

# Image translation

```python
width  = image.shape[1]
height = image.shape[0]

translation_matrix = np.array([[1,0,100],
                               [0,1,70]],
                               dtype= np.float32)

translation_image = cv2.warpAffine(image , translation_matrix , dsize=(width , height))

plt.imshow(translation_image)
```
