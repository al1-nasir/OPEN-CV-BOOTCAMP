```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

path = "../dataset_images/20151101_145049.jpg"
image  = cv2.imread(path)
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
kernel = np.ones((5, 5), np.uint8) 

eroded = cv2.erode(image , 
                   kernel, iterations= 1)
plt.imshow(eroded)
```

```python
dilated = cv2.dilate(image, 
                     kernel, iterations = 1)
plt.imshow(dilated)
```
