```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

path = "../dataset_images/20151101_145049.jpg"
image  = cv2.imread(path)
plt.imshow(image)
```

# Guassian Blur

```python
guassian = cv2.GaussianBlur(image , (7,7) , 0 )
plt.imshow(guassian)
```

# Meadian Blur

```python
median  = cv2.medianBlur(image , 5)
plt.imshow(median)
```
