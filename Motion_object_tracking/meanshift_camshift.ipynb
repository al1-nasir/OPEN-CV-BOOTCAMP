```python
import cv2
import numpy as np

cap = cv2.VideoCapture("slow.flv")

ret, frame = cap.read()

print(f"Shape is in syntax (width, height, channels) = {frame.shape}")

# Get the width and height from the frame directly
height, width = frame.shape[:2]

out = cv2.VideoWriter('car_tracking_mean_shift.avi', cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'),
                      30, (width, height))

# Initial window location
r, h, c, w = 250, 90, 400, 125  # (y, height, x, width)
track_window = (c, r, w, h)

# Setup the ROI (region of interest)
roi = frame[r:r+h, c:c+w]
hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv_roi, np.array((0., 60., 32.)), np.array((180., 255., 255.)))
roi_hist = cv2.calcHist([hsv_roi], [0], mask, [180], [0, 180])
cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)

# Termination criteria for cv2.MeanSHIFT
term_crit = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 1)

while True:
    ret, frame = cap.read()

    if ret:
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        dst = cv2.calcBackProject([hsv], [0], roi_hist, [0, 180], 1)

        ret, track_window = cv2.meanShift(dst, track_window, term_crit)

        x, y, w, h = track_window
        img = cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 255, 255), 2)
        out.write(img)
        cv2.imshow("tracking", img)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
