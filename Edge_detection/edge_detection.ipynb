```python
import cv2
import matplotlib.pyplot as plt

# Convert the image to grayscale
gray = cv2.cvtColor(browser, cv2.COLOR_BGR2GRAY)

# Apply the threshold to get a binary image
ret, thresh = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY)

# Find the contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Plot the contours
plt.figure()
for contour in contours:
    plt.plot(contour[:, 0, 0], contour[:, 0, 1], color='blue')

plt.gca().invert_yaxis()  # Invert the y-axis to match OpenCV's coordinate system
plt.show()
```
