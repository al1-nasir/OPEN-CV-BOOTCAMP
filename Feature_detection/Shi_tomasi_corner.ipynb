```python
import cv2 
import matplotlib.pyplot as plt 
path = "../dataset_images/shapes.jpg"
```

```python
img = cv2.imread(path)
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
corners  = cv2.goodFeaturesToTrack(gray , 25 , 0.01 , 10 )
corners
import numpy as np
corners = np.int0(corners)

for i in corners :
    x, y = i.ravel()
    cv2.circle(img , (x,y) , 3 , 255 )

plt.imshow(img)
```
