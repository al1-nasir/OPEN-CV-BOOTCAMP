```python
import numpy
import cv2
import matplotlib.pyplot as plt
path = "../dataset_images/shapes.jpg"
```

```python
img = cv2.imread(path)
gray  = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
harris = cv2.cornerHarris(gray , 2 , 3 , 0.04) 
dst = cv2.dilate(harris , None)
plt.imshow(harris)
```

```python
img[dst>0.01*dst.max()]=[255, 0 , 0]
plt.imshow(img)
```
